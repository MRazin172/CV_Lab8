```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/content/1.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

this is the original image

```python
figsize=((8, 8))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

Fisrst order derivatives- sobel x and y

scaherr x and y

First-order derivative methods detect edges by computing the gradient or rate of change of pixel intensities in an image.
in a grayscale image where the pixel values increase from left to right. Using the Sobel, the gradient in the horizontal direction would detect this change as an edge, indicating a transition from dark to light areas.

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)
```

Second order derivatives - laplacian

the 2nd order derivative laplacian is mainly used after applying 1st oredr derivative to enhance the edges more clearly.
Second-order derivative methods detect edges by examining the rate of change of the gradient or curvature of pixel intensities in an image.
In the same grayscale image, after detecting edges using the Sobel operator, the Laplacian operator can be applied to enhance the edges by highlighting regions where the gradient changes abruptly.

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

Optimum Edge Detection Method - canny


this is a combination of both 1st and 2nd


the optimum edge detection method aims to combine the advantages of both first and second-order derivative methods to accurately detect edges while minimizing noise and false detections.
applying the Canny edge detector to a noisy image. It would first smooth the image to reduce noise, then identify edges based on gradient changes, and finally refine the edges to obtain a clean and accurate edge map.

```python
canny = cv2.Canny(image, 100, 200)
```

Displaying the results


```python

plt.figure(figsize=(15, 10))


plt.subplot(1, 5, 1)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(scharr_x, cmap='gray')
plt.title('Scharr X')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(scharr_y, cmap='gray')
plt.title('Scharr Y')
plt.axis('off')

plt.figure(figsize=(8, 8))
plt.imshow(laplacian, cmap='gray')
plt.title('second order - Laplacian')
plt.axis('off')

plt.figure(figsize=(8, 8))
plt.imshow(canny, cmap='gray')
plt.title('optimum Edge Detection Method - Canny Edge Detection')
plt.axis('off')

plt.show()
```
